```python
import pandas as pd
import matplotlib.pyplot as plt

# Función para limpiar y procesar una línea específica del CSV
def process_line(line):
    parts = [part.strip() for part in line.split(';') if part.strip()]
    return parts

# Extraer líneas relevantes del archivo CSV
file_path = 'total_alumnos.csv'
lines = []
with open(file_path, 'r', encoding='latin1') as file:
    for line in file:
        if any(char.isdigit() for char in line):
            lines.append(process_line(line))

# Crear un DataFrame a partir de las líneas procesadas
clean_data = pd.DataFrame(lines)

# Datos para la distribución de alumnos por nivel educativo
distribution_data = clean_data.iloc[2:5, [0, 1]].copy()
distribution_data.columns = ['Nivel Educativo', 'Total Alumnos']
distribution_data['Total Alumnos'] = distribution_data['Total Alumnos'].str.replace('.', '').astype(int)

# Gráfico 1: Distribución de Alumnos por Nivel Educativo
plt.figure(figsize=(10, 6))
plt.bar(distribution_data['Nivel Educativo'], distribution_data['Total Alumnos'], color=['blue', 'green', 'red'])
plt.xlabel('Nivel Educativo')
plt.ylabel('Total de Alumnos')
plt.title('Distribución de Alumnos por Nivel Educativo en CABA (2022)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Limpiar los valores nulos reemplazándolos por 0 para la conversión a enteros
comparison_data = clean_data.iloc[1, 2:12].replace({None: '0'}).str.replace('.', '').astype(int)

# Reindexar con los nombres de las categorías
comparison_data.index = ['Total', 'Estatal Inicial', 'Privada Inicial', 'Estatal Primario', 'Privada Primario', 'Estatal Secundario', 'Privada Secundario', 'Total Estatal', 'Total Privado', 'Total General']
comparison_data = comparison_data[['Estatal Inicial', 'Privada Inicial', 'Estatal Primario', 'Privada Primario', 'Estatal Secundario', 'Privada Secundario']]

# Gráfico 2: Comparación de Alumnos entre Gestión Estatal y Privada
plt.figure(figsize=(12, 7))
comparison_data.plot(kind='bar', stacked=True, color=['blue', 'orange', 'green', 'red', 'purple', 'brown'], figsize=(12, 7))
plt.xlabel('Nivel Educativo')
plt.ylabel('Total de Alumnos')
plt.title('Comparación de Alumnos entre Gestión Estatal y Privada en CABA (2022)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Crear un DataFrame con los totales de alumnos por nivel para el gráfico comparativo estático
static_comparison_data = distribution_data.set_index('Nivel Educativo').copy()

# Gráfico 3: Número de Alumnos en los Diferentes Niveles Educativos
plt.figure(figsize=(10, 6))
static_comparison_data.plot(kind='line', marker='o', figsize=(10, 6))
plt.xlabel('Nivel Educativo')
plt.ylabel('Total de Alumnos')
plt.title('Número de Alumnos en los Diferentes Niveles Educativos en CABA (2022)')
plt.grid(True)
plt.show()

# Análisis y conclusiones
from IPython.display import display, Markdown

display(Markdown("### Análisis del Gráfico 1"))
display(Markdown("La distribución de alumnos muestra que la mayoría de los estudiantes están matriculados en el nivel primario, seguido del nivel secundario y finalmente del nivel inicial. Esto es coherente con la estructura educativa típica, donde los estudiantes pasan más tiempo en los niveles primarios y secundarios."))

display(Markdown("### Análisis del Gráfico 2"))
display(Markdown("El gráfico muestra que la mayoría de los alumnos están en instituciones de gestión estatal en comparación con las privadas, tanto en el nivel inicial como en el primario y secundario. La diferencia es más notable en los niveles primario y secundario."))

display(Markdown("### Análisis del Gráfico 3"))
display(Markdown("El gráfico comparativo estático reafirma que el nivel primario tiene la mayor cantidad de alumnos, seguido por el nivel secundario y luego el nivel inicial. Esto proporciona una visión clara de la distribución de los estudiantes en los diferentes niveles educativos."))

display(Markdown("### Conclusiones Generales"))
display(Markdown("1. **Distribución de Alumnos por Nivel Educativo**: La mayor cantidad de alumnos se encuentra en el nivel primario, seguido del nivel secundario y el nivel inicial."))
display(Markdown("2. **Comparación entre Gestión Estatal y Privada**: Las instituciones de gestión estatal tienen una mayor cantidad de alumnos en todos los niveles educativos en comparación con las privadas."))
display(Markdown("3. **Evolución del Número de Alumnos en los Diferentes Niveles**: La comparación estática confirma que el nivel primario domina en términos de matrícula de estudiantes, seguido por el secundario y el inicial."))
display(Markdown("### Conclusión General"))
display(Markdown("El análisis de los datos revela que la distribución de alumnos en la Ciudad Autónoma de Buenos Aires está fuertemente concentrada en las instituciones estatales, especialmente en el nivel primario. La estructura educativa refleja una mayor cantidad de alumnos en el nivel primario, lo cual es típico, y esta tendencia se mantiene en las instituciones tanto estatales como privadas."))
```
