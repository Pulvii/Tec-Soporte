```python
import pandas as pd
import matplotlib.pyplot as plt

# Funcin para limpiar y procesar una lnea especfica del CSV
def process_line(line):
    parts = [part.strip() for part in line.split(';') if part.strip()]
    return parts

# Extraer lneas relevantes del archivo CSV
file_path = 'total_alumnos.csv'
lines = []
with open(file_path, 'r', encoding='latin1') as file:
    for line in file:
        if any(char.isdigit() for char in line):
            lines.append(process_line(line))

# Crear un DataFrame a partir de las lneas procesadas
clean_data = pd.DataFrame(lines)

# Datos para la distribucin de alumnos por nivel educativo
distribution_data = clean_data.iloc[2:5, [0, 1]].copy()
distribution_data.columns = ['Nivel Educativo', 'Total Alumnos']
distribution_data['Total Alumnos'] = distribution_data['Total Alumnos'].str.replace('.', '').astype(int)

# Grfico 1: Distribucin de Alumnos por Nivel Educativo
plt.figure(figsize=(10, 6))
plt.bar(distribution_data['Nivel Educativo'], distribution_data['Total Alumnos'], color=['blue', 'green', 'red'])
plt.xlabel('Nivel Educativo')
plt.ylabel('Total de Alumnos')
plt.title('Distribucin de Alumnos por Nivel Educativo en CABA (2022)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Limpiar los valores nulos reemplazndolos por 0 para la conversin a enteros
comparison_data = clean_data.iloc[1, 2:12].replace({None: '0'}).str.replace('.', '').astype(int)

# Reindexar con los nombres de las categoras
comparison_data.index = ['Total', 'Estatal Inicial', 'Privada Inicial', 'Estatal Primario', 'Privada Primario', 'Estatal Secundario', 'Privada Secundario', 'Total Estatal', 'Total Privado', 'Total General']
comparison_data = comparison_data[['Estatal Inicial', 'Privada Inicial', 'Estatal Primario', 'Privada Primario', 'Estatal Secundario', 'Privada Secundario']]

# Grfico 2: Comparacin de Alumnos entre Gestin Estatal y Privada
plt.figure(figsize=(12, 7))
comparison_data.plot(kind='bar', stacked=True, color=['blue', 'orange', 'green', 'red', 'purple', 'brown'], figsize=(12, 7))
plt.xlabel('Nivel Educativo')
plt.ylabel('Total de Alumnos')
plt.title('Comparacin de Alumnos entre Gestin Estatal y Privada en CABA (2022)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Crear un DataFrame con los totales de alumnos por nivel para el grfico comparativo esttico
static_comparison_data = distribution_data.set_index('Nivel Educativo').copy()

# Grfico 3: Nmero de Alumnos en los Diferentes Niveles Educativos
plt.figure(figsize=(10, 6))
static_comparison_data.plot(kind='line', marker='o', figsize=(10, 6))
plt.xlabel('Nivel Educativo')
plt.ylabel('Total de Alumnos')
plt.title('Nmero de Alumnos en los Diferentes Niveles Educativos en CABA (2022)')
plt.grid(True)
plt.show()

# Anlisis y conclusiones
from IPython.display import display, Markdown

display(Markdown("### Anlisis del Grfico 1"))
display(Markdown("La distribucin de alumnos muestra que la mayora de los estudiantes estn matriculados en el nivel primario, seguido del nivel secundario y finalmente del nivel inicial. Esto es coherente con la estructura educativa tpica, donde los estudiantes pasan ms tiempo en los niveles primarios y secundarios."))

display(Markdown("### Anlisis del Grfico 2"))
display(Markdown("El grfico muestra que la mayora de los alumnos estn en instituciones de gestin estatal en comparacin con las privadas, tanto en el nivel inicial como en el primario y secundario. La diferencia es ms notable en los niveles primario y secundario."))

display(Markdown("### Anlisis del Grfico 3"))
display(Markdown("El grfico comparativo esttico reafirma que el nivel primario tiene la mayor cantidad de alumnos, seguido por el nivel secundario y luego el nivel inicial. Esto proporciona una visin clara de la distribucin de los estudiantes en los diferentes niveles educativos."))

display(Markdown("### Conclusiones Generales"))
display(Markdown("1. **Distribucin de Alumnos por Nivel Educativo**: La mayor cantidad de alumnos se encuentra en el nivel primario, seguido del nivel secundario y el nivel inicial."))
display(Markdown("2. **Comparacin entre Gestin Estatal y Privada**: Las instituciones de gestin estatal tienen una mayor cantidad de alumnos en todos los niveles educativos en comparacin con las privadas."))
display(Markdown("3. **Evolucin del Nmero de Alumnos en los Diferentes Niveles**: La comparacin esttica confirma que el nivel primario domina en trminos de matrcula de estudiantes, seguido por el secundario y el inicial."))
display(Markdown("### Conclusin General"))
display(Markdown("El anlisis de los datos revela que la distribucin de alumnos en la Ciudad Autnoma de Buenos Aires est fuertemente concentrada en las instituciones estatales, especialmente en el nivel primario. La estructura educativa refleja una mayor cantidad de alumnos en el nivel primario, lo cual es tpico, y esta tendencia se mantiene en las instituciones tanto estatales como privadas."))
```
