```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = 'Preguntas_Revisadas_sobre_Apuestas_Deportivas_en_Adolescentes.csv'
data = pd.read_csv(file_path)

# Renombrar las columnas y transponer el DataFrame
data.columns = ['Pregunta'] + [f'Participante_{i}' for i in range(1, len(data.columns))]
data_transposed = data.set_index('Pregunta').T

# Codificacin de respuestas
data_transposed.replace({
    'S': 1,
    'No': 0,
    'Nunca': 0,
    'Rara vez': 1,
    'A veces': 2,
    'Frecuentemente': 3,
    'Siempre': 4,
    'Menos de $10': 1,
    '$10 - $50': 2,
    '$50 - $100': 3,
    'Ms de $100': 4
}, inplace=True)



# Grficos y Anlisis

# Frecuencia de participacin en apuestas deportivas
frecuencia = data_transposed['Con qu frecuencia realizas apuestas deportivas?'].value_counts().sort_index()

# Grfico de barras para la frecuencia de participacin
plt.figure(figsize=(10, 6))
frecuencia.plot(kind='bar')
plt.title('Frecuencia de Participacin en Apuestas Deportivas')
plt.xlabel('Frecuencia')
plt.ylabel('Cantidad de Participantes')
plt.xticks(rotation=0)
plt.show()

# Gasto mensual en apuestas deportivas
gasto = data_transposed['Cunto dinero sueles gastar en apuestas deportivas al mes?'].value_counts().sort_index()

# Grfico de barras para el gasto mensual
plt.figure(figsize=(10, 6))
gasto.plot(kind='bar')
plt.title('Gasto Mensual en Apuestas Deportivas')
plt.xlabel('Monto ($)')
plt.ylabel('Cantidad de Participantes')
plt.xticks(rotation=0)
plt.show()

# Percepcin sobre el impacto negativo
impacto_negativo = data_transposed['Te preocupa el impacto de las apuestas deportivas en tu vida diaria?'].value_counts()

# Grfico de pastel para la percepcin sobre el impacto negativo
plt.figure(figsize=(8, 8))
impacto_negativo.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Percepcin sobre el Impacto Negativo de las Apuestas Deportivas')
plt.ylabel('')
plt.show()


```
